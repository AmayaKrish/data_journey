```python
import pandas as pd
import numpy as np

data = {
    "Patient_ID": [1, 2, 3, 4, 4],
    "Age": [25, 40, np.nan, 60, 60],
    "Blood_Pressure": [120, 140, 130, None, 145],
    "Diabetes": ["No", "Yes", "No", "Yes", "Yes"]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df["Blood_Pressure"] = df["Blood_Pressure"].fillna(df["Blood_Pressure"].median())
```
